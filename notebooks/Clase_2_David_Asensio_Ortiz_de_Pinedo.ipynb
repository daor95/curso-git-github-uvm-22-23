# Excepciones

En Python, los errores de ejecución de código se denominan excepciones. Dichas excepciones, por supuesto, son clases derivadas de una clase de excepción original, `Exception`. Sus variantes incluyen, entre otras:

* `ZeroDivisionError`: generada por la división entre `0`.
* `IndexError`: generada por intentos de acceso a índices sobre la longitud total de una estructura secuencial.
* `NameError`: generada por la llamada a una clase, función o variable no definida previamente. 

## Traceback

El _traceback_ es el registro de llamadas que indica exactamente dónde tiene lugar una excepción determinada. Es importante saber leer adecuadamente el _traceback_ para poder determinar el error verdadero surgido en el código.



## Manejo de excepciones

El manejo de excepciones hace referencia a la técnica empleada para gestionar posibles excepciones en el código y corregirlas, _debuggearlas_, silenciarlas, etc.

La estructura en Python consiste, cómo no, en una serie de bloques indentados encabezados por las palabras reservadas `try`, `except`, `else` y `finally`.

```python
# Without exception catch:

def divide(value_1, value_2):
  return value_1 / value_2


print(divide(4, 2), end="\n\n")
print(divide(2, 0))
```

```python
# With exception catch:

def divide(value_1, value_2):
  output = None
  
  try:
    output = value_1 / value_2
  except ZeroDivisionError:
    print("Warning: The second value cannot be zero.")

  return output


print(divide(4, 2), end="\n\n")
print(divide(2, 0))
```

```python
# Exception catch with else statement:

def divide(value_1, value_2):
  output = None

  try:
    output = value_1 / value_2
  except ZeroDivisionError:
    print("Warning: The second value cannot be zero.")
  else:
    print("The calculation was OK.")

  return output


print(divide(4, 2), end="\n\n")
print(divide(2, 0))
```

```python
# Exception catch with else statement:

def divide(value_1, value_2):
  output = None
  
  try:
    output = value_1 / value_2
  except ZeroDivisionError:
    print("Warning: The second value cannot be zero.")
  else:
    print("The calculation was OK.")
  finally:
    print("Finished calculation process.")

  return output


print(divide(4, 2), end="\n\n")
print(divide(2, 0))
```

Los arrays de Numpy son la versión optimizada de las listas estándar de Python. De hecho, toda lista se puede convertir en array de la siguiente manera:

```python
import numpy as np

# Data conversion:

list_1 = [1, 2, 3, 4]  # Standard list.
array_1 = np.array(list_1)  # Numpy array.

# Demonstration:

print("list_1 * 2 = ", list_1 * 2)
print("array_1 * 2 = ", array_1 * 2)
```

# SciPy

SciPy es una librería que permite realizar operaciones de cálculo científico de manera controlada y rápida. Podría entenderse que Numpy representa las estructuras matemáticas y realiza las operaciones internas, mientras que SciPy se encarga de utilizar todo eso y ofrecer funcionalidades complejas como derivación, integración, interpolación, etc.

```python
import numpy as np
from scipy.interpolate import interp1d
from random import randrange

# Original x and y values:

x_values = np.arange(0, 20)
y_values = [randrange(0, 10) for x in x_values]  # Randomly generated y values.

# Interpolation function:

interpolation_function = interp1d(x_values, y_values) 

# New x values (more precise) and interpolated y values:

x_values_interp = np.arange(0, 19, .25)
y_values_interp = interpolation_function(x_values_interp)

plt.plot(x_values_interp, y_values_interp, '.-', color = "darkorange", 
         label = "Interpolated data")
plt.plot(x_values, y_values, '.', color = "red", label = "Raw data")

plt.title("Raw data vs. interpolated data")
plt.legend()
plt.grid(True)

plt.show()
```

Calcula la integral de la función `f(x) = sin(x)` en el intervalo `[0, π / 2]` mediante la suma de Riemman y mediante SciPy. Muestra en pantalla la diferencia de valores entre ambos resultados.

También deberás medir los tiempos de ejecución de cada método mediante la función `perf_counter` de la librería `time`.

```python
# Write your code below:

#Import Modules
import time
import numpy as np
import math
import scipy.integrate as integrate


#Riemman

def calculate_integral(function, lower, upper, number_trapezoids) -> float:
    """Riemman Integral Calculation

    It takes as input the base function, the lower and upper limits of the
    interval for (x) and the number of trapezoids in that range in order
    to calculate the Riemman Integral. The output is a float number resulting of
    the calculation of the Riemman Integral.
    """
    base = (upper - lower)/number_trapezoids
    result = 0.5*function(lower) + sum([function(lower + i*base) for i in range(1, 
    number_trapezoids)]) + 0.5*function(upper)
    result *= base
    return result


#Basic values for x and calculation of number of trapezoids.
x_values = np.arange(0, math.pi / 2, 0.2)
number_trapezoids = int((math.pi / 2) / 0.2)

t1_start = time.perf_counter() # Time Counter Start

result = calculate_integral(math.sin, 0, math.pi / 2, number_trapezoids)
print("\n\n",result) #Integral Calculation from Function

t1_stop = time.perf_counter() #Time Counter Stop

print("\n\nTime elapsed in seconds (Riemann): ",t1_stop - t1_start)
#Total Time Elapsed Compute



#SCIPY

t2_start = time.perf_counter() #Time Counter Start

x2 = lambda x: math.sin(x) #Define x to be used as a variable of the function

result = integrate.quad(x2, 0, math.pi / 2) #Integral Calculation
print ("\n\n",result) #Output: result + coeificient of error

t2_stop = time.perf_counter() #Time Counter Stop

print("\n\nTime elapsed in seconds (SCIPY): ",t2_stop - t2_start)
#Total Time Elapsed Compute
```

## _Ejercicio: Tratamiento de imagen_

Sabiendo que la imagen del ejemplo anterior tiene 3 canales de color, mostrar tres descomposiciones de la misma en pantalla. En cada una de ellas deben verse solo los canales rojo, verde y azul, respectivamente.

```python
# Write your code below:

#Import Modules
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np

#Obtain Sample Image
img = misc.face()  # Sample image

#Get information of the Image
def image_data(img):
  """Get Basic Info from Image

  Print Information about the type of the image data, number of channels (number
  of colors and dimension (resolution) of the image.
  """

  print("Data type of image data: ", type(img)) #Data Type
  print("Image channels: ", img.ndim) #Number of channels (colors)
  print("Image dimensions: ", img.shape) #Image Dimension/Size/Resolution



def remove_colors(img):
  """Remove Colors and Show Image.

  Remove colors indicating which channels have been removed and output new image.
  This works for all kind of images with independence of the number of channels.
  """

  for dim in range(img.ndim):
    gen_list = [i for i in range(img.ndim) if i!=dim] #Generate List with the
    #Channels that must be set to 0.
    print("\nChannels set as 0: ", gen_list) #Print which channels are set to 0.
    new_image = np.copy(img) #Create a copy of the original image.

    for x in gen_list:
      new_image[:, :, x]=0 #Set the previously defined channels to 0.
  
    plt.imshow(new_image)  #Plot and show image output.
    plt.show()

image_data(img)
remove_colors(img)
```
